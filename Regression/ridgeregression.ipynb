```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import make_scorer
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv('E:/Ramya/brushUps/Basics/ML using sklearn/Projects/Wine_quality/data/winequality-white.csv',delimiter=';')
```

```python
X=data.iloc[:,0:11]
y=data.iloc[:,11]
```

```python
col_names=X.columns
scaler=preprocessing.MinMaxScaler()
normalized_list=scaler.fit_transform(X)
X=pd.DataFrame(normalized_list,columns=col_names)
```

```python
r2=make_scorer(r2_score)
```

```python
grid_params={'alpha':[1e-15,1e-10,1e-8,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
clf=GridSearchCV(Ridge(),param_grid=grid_params,cv=10,return_train_score=False,scoring=r2)
clf.fit(X,y)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
ridgereg=Ridge(alpha=clf.best_params_['alpha'])
y_pred=cross_val_predict(ridgereg,X,y,cv=10)
```

```python
r2_value=r2_score(y,y_pred)
adj_r2=1-((1-r2_value)*(X.shape[0]-1)/(X.shape[0]-X.shape[1]-1))
```
